```python
# prompt: plotting hyperbola code

import matplotlib.pyplot as plt
import numpy as np

# Define the range for x values
x = np.linspace(-5, 5, 400)

# Define the hyperbola equation (example: x^2/4 - y^2/9 = 1)
# Solve for y: y = Â±sqrt(9 * (x^2/4 - 1))
y_positive = np.sqrt(9 * (x**2 / 4 - 1))
y_negative = -np.sqrt(9 * (x**2 / 4 - 1))


# Create the plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.plot(x, y_positive, label="y = +sqrt(9 * (x^2/4 - 1))", color='blue')
plt.plot(x, y_negative, label="y = -sqrt(9 * (x^2/4 - 1))", color='red')

# Customize the plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hyperbola Plot")
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k')  # Add x-axis
plt.axvline(x=0, color='k')  # Add y-axis

# Set limits for better visualization
plt.xlim([-5, 5])
plt.ylim([-6, 6])  # Adjust y-limits as needed

# Show the plot
plt.show()
```

```python
# prompt: plotting y^2=16x code

import matplotlib.pyplot as plt
import numpy as np

# Define the range for x values
x = np.linspace(0, 5, 400)  # x values must be greater than or equal to 0

# Calculate y values for the parabola y^2 = 16x
y_positive = np.sqrt(16 * x)
y_negative = -np.sqrt(16 * x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y_positive, label="y = +sqrt(16x)", color='blue')
plt.plot(x, y_negative, label="y = -sqrt(16x)", color='red')

# Customize the plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Parabola y^2 = 16x")
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k')  # Add x-axis
plt.axvline(x=0, color='k')  # Add y-axis


# Show the plot
plt.show()
```
